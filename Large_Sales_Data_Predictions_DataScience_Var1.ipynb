```python
import os
import pandas as pd
```

```python
# 1 data import from all excel files, read and merge as one file


path = "./Sales_Data"


files =  [file for file in os.listdir(path) if not file.startswith('.')]

all_months_data = pd.DataFrame()

for file in files:
    Excel_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data,Excel_data])

all_months_data.to_csv("all_data_Copy.csv",index=False)


```
